# GZ2 decision tree

In the next table we can see the GZ2 decision tree, comprising 11 tasks and 37 responses. The 'Task' number is an abbreviation only and does not necessarily represent the order of the task within the decision tree.

The text in 'Question' and 'Responses' are displayed to volunteers during classification. 'Next' gives the subsequent task for the chosen response.

The GZ2 tree has 11 classification tasks with a total of 37 possible responses. A classifier selects only one response for each task, after which they are immediately taken to the next task in the tree.
